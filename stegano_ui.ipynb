```python
from stegano import lsb
import tkinter as tk
```

```python
def encrypt_xor(message, key):
  """
  Fungsi untuk mengenkripsi pesan dengan XOR.

  Args:
    message: Pesan yang akan dienkripsi (str).
    key: Kunci untuk enkripsi (str).

  Returns:
    str: Pesan terenkripsi.
  """
  message = message.upper()
  key = key.upper()

#   if len(message) != len(key):
#     print("Panjang kunci harus sama dengan teks.")
#     return

  encrypted_text = ''

  for i in range(len(message)):
    if message[i].isalpha():
      message_char = ord(message[i]) - ord('A')
      key_char = ord(key[i]) - ord('A')

      # Menggunakan operasi XOR
      encrypted_char = message_char ^ key_char
      encrypted_text += chr(encrypted_char + ord('A'))
    else:
      encrypted_text += message[i]

  return encrypted_text

def decrypt_xor(encrypted_text, key):
  """
  Fungsi untuk dekripsi pesan dengan XOR.

  Args:
    encrypted_text: Pesan terenkripsi (str).
    key: Kunci untuk dekripsi (str).

  Returns:
    str: Pesan yang telah didekripsi.
  """
  encrypted_text = encrypted_text.upper()
  key = key.upper()

#   if len(encrypted_text) != len(key):
#     print("Panjang kunci harus sama dengan teks.")
#     return

  decrypted_text = ''

  for i in range(len(encrypted_text)):
    if encrypted_text[i].isalpha():
      encrypted_char = ord(encrypted_text[i]) - ord('A')
      key_char = ord(key[i]) - ord('A')

      # Menggunakan operasi XOR untuk dekripsi
      decrypted_char = encrypted_char ^ key_char
      decrypted_text += chr(decrypted_char + ord('A'))
    else:
      decrypted_text += encrypted_text[i]

  return decrypted_text

def main():
  window = tk.Tk()
  window.title("Enkripsi dan Dekripsi Steganografi")

  # Label dan entry untuk plaintext
  plaintext_label = tk.Label(window, text="Plaintext:")
  plaintext_label.grid(row=0, column=0)
  plaintext_entry = tk.Entry(window, width=50)
  plaintext_entry.grid(row=0, column=1, columnspan=2)

  # Label dan entry untuk key
  key_label = tk.Label(window, text="Key:")
  key_label.grid(row=1, column=0)
  key_entry = tk.Entry(window, width=50)
  key_entry.grid(row=1, column=1, columnspan=2)

  # Label dan entry untuk output
  output_label = tk.Label(window, text="Output:")
  output_label.grid(row=3, column=0)
  output_entry = tk.Entry(window, width=50)
  output_entry.grid(row=3, column=1, columnspan=2)

  # Fungsi untuk enkripsi dan sembunyi
  def encrypt_and_hide(plaintext, key):
    if len(plaintext) != len(key):
      output_entry.delete(0, tk.END)
      output_entry.insert(0, "Panjang teks harus sama dengan kunci.")
      return
    encrypted_message = encrypt_xor(plaintext, key)
    secret = lsb.hide("Sc.png", encrypted_message)  # Pastikan path file ini benar
    secret.save("Sc-sec.png")
    output_entry.delete(0, tk.END)
    output_entry.insert(0, "Pesan terenkripsi: " + encrypted_message)

  # Fungsi untuk dekripsi dan tampilkan
  def decrypt_and_show(key):
    decrypted_message = lsb.reveal("Sc-sec.png")
    if decrypted_message is None:
      output_entry.delete(0, tk.END)
      output_entry.insert(0, "Tidak ada pesan terenkripsi dalam gambar.")
      return
    if len(decrypted_message) != len(key):
      output_entry.delete(0, tk.END)
      output_entry.insert(0, "Panjang kunci harus sama dengan teks.")
      return
    decrypted_text = decrypt_xor(decrypted_message, key)
    output_entry.delete(0, tk.END)
    output_entry.insert(0, "Teks terdekripsi: " + decrypted_text)

  # Mengubah perintah tombol
  encrypt_button = tk.Button(window, text="Enkripsi dan Sembunyikan", command=lambda: encrypt_and_hide(plaintext_entry.get(), key_entry.get()))
  encrypt_button.grid(row=2, column=0)
  decrypt_button = tk.Button(window, text="Dekripsi dan Tampilkan", command=lambda: decrypt_and_show(key_entry.get()))
  decrypt_button.grid(row=2, column=1)

  window.mainloop()

if __name__ == "__main__":
  main()
```
